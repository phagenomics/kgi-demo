# Lab 1. DotPlot Matrix for sequence alignment
A dot plot is a graphical reprensentation of the similarity between two sequences, it is very simple, intuitive and a good way to visualiza aligments between two sequences.
It is done by plotting two sequences against each other in a two dimensional grid, each position in the grid corresponds to a base pair position in each sequence, if they match, ie. have the same base, a dot is plotted. 
Visual inspection will show diagonals in regions of similarity.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

#store inputs in two variables
seq1 = input("Enter the first sequence: ")
seq2 = input("Enter the second sequence: ")

#create a zeros matrix with dimensions equal to the length of the sequences
matrix = np.zeros((len(seq1), len(seq2)))

# compare each sequence, make the cell equal to one if sequences match each other
for i in range(len(seq1)):
    for j in range(len(seq2)):
        if seq1[i] == seq2[j]:
            matrix[i, j] = 1

#plot matrix as scatter plot
fig, ax = plt.subplots(figsize=(len(seq2)/6, len(seq1)/6))
ax.scatter(*np.where(matrix==1), s=40, marker='o', color='blue')

ax.set_xticks(np.arange(len(seq1)))
ax.set_yticks(np.arange(len(seq2)))



ax.set_xticklabels(list(seq1))
ax.set_yticklabels(list(seq2))

ax.set_xlabel("Sequence 1")
ax.set_ylabel("Sequence 2")

plt.show()
```
